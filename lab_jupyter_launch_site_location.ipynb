```python
## Task 1: Mark all launch sites on a map
# Create a Folium map marking SpaceX launch sites and save it as an HTML file

import folium

# Known SpaceX launch sites and their coordinates
launch_sites = [
    {"name": "CCAFS LC-40 (Cape Canaveral, Florida)", "lat": 28.5618571, "lon": -80.577366},
    {"name": "KSC LC-39A (Kennedy Space Center, Florida)", "lat": 28.6080585, "lon": -80.6039558},
    {"name": "VAFB SLC-4E (Vandenberg, California)", "lat": 34.632093, "lon": -120.610829},
]

# Calculate a simple center for the initial map view
avg_lat = sum(site["lat"] for site in launch_sites) / len(launch_sites)
avg_lon = sum(site["lon"] for site in launch_sites) / len(launch_sites)

# Create the base map
m = folium.Map(location=[avg_lat, avg_lon], zoom_start=4, tiles="OpenStreetMap")

# Add markers for each launch site
for site in launch_sites:
    folium.Marker(
        location=[site["lat"], site["lon"]],
        popup=f"{site['name']}<br>Lat: {site['lat']}, Lon: {site['lon']}",
        tooltip=site["name"],
        icon=folium.Icon(icon="rocket", prefix="fa")
    ).add_to(m)
    
    # Optional: draw a small circle around each site for emphasis
    folium.Circle(
        location=[site["lat"], site["lon"]],
        radius=500,
        fill=True,
        fill_opacity=0.2,
        opacity=0.5
    ).add_to(m) 
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
import math

def haversine_distance(lat1, lon1, lat2, lon2):
    """
    Calculate distance between two lat/long points (in kilometers).
    """
    R = 6371  # Earth radius in km
    
    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    
    # Differences
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    # Haversine formula
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))
    
    return R * c  # distance in km

# Example: distance between CCAFS LC-40 and KSC LC-39A
dist = haversine_distance(28.5618571, -80.577366, 28.6080585, -80.6039558)
print(f"Distance: {dist:.2f} km") 
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance 
```
